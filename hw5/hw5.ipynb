```python
from python.svm import *
from python.svmutil import *
import pandas as pd
import numpy as np
from numpy import savetxt
import csv
import os
```

```python
count_exp = 0
count_a, count_b, count_c, count_d, count_e = 0, 0, 0, 0, 0

for i in range(1000):

    df = pd.read_csv('train_6.txt', header=None, sep='\t').to_numpy()

    d = dict()
    select_list = []
    count = 0

    while count < 200:
        i = np.random.randint(0, len(df))
        if (i not in d):
            d[i] = 1
            count += 1
            select_list.append(i)

    df_list = list(range(0, len(df)))
    for element in select_list:
        df_list.remove(element)


    if os.path.isfile('train_6_tmp.txt'):
        os.remove('train_6_tmp.txt')
    if os.path.isfile('test_6_tmp.txt'):
        os.remove('test_6_tmp.txt')

    # train_data
    for element in select_list:
        if df[element][0][0] == '1':        
            with open('train_6_tmp.txt','a') as f:
                writer = csv.writer(f)
                f.write(str(1) + ' ' + str(df[element][0][2:]))
                f.write("\n")
        else:
            with open('train_6_tmp.txt','a') as f:
                writer = csv.writer(f)
                f.write(str(-1) + ' ' + str(df[element][0][2:]))
                f.write("\n")

    # test_data
    for element in df_list:
        if df[element][0][0] == '1':        
            with open('test_6_tmp.txt','a') as f:
                writer = csv.writer(f)
                f.write(str(1) + ' ' + str(df[element][0][2:]))
                f.write("\n")
        else:        
            with open('test_6_tmp.txt','a') as f:
                writer = csv.writer(f)
                f.write(str(-1) + ' ' + str(df[element][0][2:]))
                f.write("\n")

    gamma = 0.1

    y, x = svm_read_problem('train_6_tmp.txt')
    yt, xt = svm_read_problem('test_6_tmp.txt')
    param = '-s 0 -t 2 -g ' + str(gamma) + ' -c 0.1'
    model = svm_train(y, x, param)

    p_label_0, p_acc_0, p_val_0 = svm_predict(yt, xt, model)

    gamma = 1

    y, x = svm_read_problem('train_6_tmp.txt')
    yt, xt = svm_read_problem('test_6_tmp.txt')
    param = '-s 0 -t 2 -g ' + str(gamma) + ' -c 0.1'
    model = svm_train(y, x, param)

    p_label_1, p_acc_1, p_val_1 = svm_predict(yt, xt, model)

    gamma = 10

    y, x = svm_read_problem('train_6_tmp.txt')
    yt, xt = svm_read_problem('test_6_tmp.txt')
    param = '-s 0 -t 2 -g ' + str(gamma) + ' -c 0.1'
    model = svm_train(y, x, param)

    p_label_2, p_acc_2, p_val_2 = svm_predict(yt, xt, model)

    gamma = 100

    y, x = svm_read_problem('train_6_tmp.txt')
    yt, xt = svm_read_problem('test_6_tmp.txt')
    param = '-s 0 -t 2 -g ' + str(gamma) + ' -c 0.1'
    model = svm_train(y, x, param)

    p_label_3, p_acc_3, p_val_3 = svm_predict(yt, xt, model)

    gamma = 1000

    y, x = svm_read_problem('train_6_tmp.txt')
    yt, xt = svm_read_problem('test_6_tmp.txt')
    param = '-s 0 -t 2 -g ' + str(gamma) + ' -c 0.1'
    model = svm_train(y, x, param)

    p_label_4, p_acc_4, p_val_4 = svm_predict(yt, xt, model)

    minmax_list = [p_acc_0, p_acc_1, p_acc_2, p_acc_3, p_acc_4]
    index = minmax_list.index(max(minmax_list))

    if index == 0:
        count_a += 1
    if index == 1:
        count_b += 1
    if index == 2:
        count_c += 1
    if index == 3:
        count_d += 1
    if index == 4:
        count_e += 1
    
    count_exp += 1
    print ('experiment : ' + str(count_exp),', a = ' + str(count_a), 'b = ' + str(count_b), 'c = ' + str(count_c), 'd = ' + str(count_d), 'e = ' + str(count_e))
```
